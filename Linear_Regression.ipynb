```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/homeprices.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('Area (Square feet)')
plt.ylabel('Price (US Dollars)')
plt.scatter(df.area, df.price, color = 'red', marker = '+')
```

```python
#linear reg object
reg = linear_model.LinearRegression()
reg.fit(df[['area']], df.price) #Training linear regression model
```

```python
reg.predict([[3300]])
```

```python
reg.coef_ #gradient or slope = m, y = mx + b
```

```python
reg.intercept_ #b = intercept
```

```python
y = 135.78767123 * 3300 + 180616.43835616432 # y = mx + b
y
```

```python
%matplotlib inline
plt.xlabel('Area', fontsize = 20)
plt.ylabel('Price', fontsize = 20)
plt.scatter(df.area, df.price, color = 'red', marker = '+')
plt.plot(df.area, reg.predict(df[['area']]), color = 'blue')
```

```python
reg.predict([[5000]])
```

```python
d = pd.read_csv('/content/areas.csv')
d.head(3)
```

```python
p = reg.predict(d)
```

```python
d['prices'] = p # create a new column called 'prices' and store the predictions in it
```

```python
d.to_csv('predictions.csv', index = False) #prediction file create
```
