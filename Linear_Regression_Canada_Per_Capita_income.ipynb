```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('/content/canada_per_capita_income.csv')
df.head()
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Income (US Dollars)')
plt.scatter(df.year, df['per capita income (US$)'], color = 'red', marker = '*')
```

```python
#linear reg object
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df['per capita income (US$)']) #Training linear regression model
```

```python
reg.predict([[1985]])
```

```python
reg.coef_ #gradient or slope = m, y = mx + b
```

```python
reg.intercept_ #b = intercept
```

```python
y = reg.coef_ * 1985 + reg.intercept_
y
```

```python
%matplotlib inline
plt.xlabel('Year', fontsize = 20)
plt.ylabel('Income', fontsize = 20)
plt.scatter(df.year, df['per capita income (US$)'], color = 'red', marker = '*')
plt.plot(df.year, reg.predict(df[['year']]), color = 'blue')
```

```python
reg.predict([[2200]])
```
